# Exploratory Data Analysis


```python
df.columns
```

**Shipping Status Distribution**

```python
# counts shipping status
Status_counts = df['Status'].value_counts()

# Plot bar Chart
plt.figure (figsize = (15,11))
Status_counts.plot (kind = 'bar', color = 'skyblue', edgecolor = 'black')

plt.title ('Shipping Status Distribution')
plt.xlabel ('Shipping Status')
plt.ylabel ('counts')
plt.xticks (rotation = 45)
plt.tight_layout()
plt.show()
```

```python
# Pie chart for shipping distribution

# count shipping status
status_counts = df['Status'].value_counts()

# Plot pie chart
plt.figure(figsize = (6,6))
status_counts.plot (kind = 'pie', autopct = '%1.1f%%', startangle = 90, 
                   color = ['skyblue', 'lightgreen', 'salmon'],
                   wedgeprops = {'edgecolor' : 'black'})
plt.title('Shipping Status Distribution', fontsize = 14)
plt.ylabel('') # remove y - label for clean look
plt.show()
```

---
###  **Key Insights:**

* **Shipped - Delivered to Buyer** dominates the chart at **76.6%**, showing that the **majority of orders are successfully delivered**.
* **Cancelled orders** make up **13.7%**, which is relatively high and may require further investigation.
* **Returned to Seller** accounts for **5.2%**, suggesting a moderate rate of returns.
* Other statuses (e.g., Picked Up, Pending, Out for Delivery, Lost in Transit) are **below 1%** each and have **minimal impact** on overall shipping outcomes.

---

### **Conclusion:**

The fulfillment process is generally effective, but there is a **notable cancellation and return rate (~19%)** that could indicate **customer issues, product dissatisfaction, or delivery problems**—areas worth exploring for improvement.

# Group By
#### The groupby() function in pandas is used to group data based on one or more columns in a DataFrame

```python
df.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)
```

```python
S_Qty=df.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)

sns.barplot(x='Size',y='Quantity', data=S_Qty)
plt.show()
```

##### Note: From above Graph you can see that most of the Qty buys M-Size in the sales
